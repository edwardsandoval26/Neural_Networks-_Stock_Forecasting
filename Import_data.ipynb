```python
markets = ["FB","AMZN","NFLX","AAPL","MSFT","GOOGL","TWTR","UBER","NVDA","BABA"] 
#FACEBOOK, AMAZON, NETFLIX, APPLE, MICROSOFT, GOOGLE, TWITTER, UBER, NVIDIA, alibaba
slices = ["year1month1","year1month2","year1month3","year1month4","year1month5","year1month6"
          "year1month7","year1month8","year1month9","year1month10","year1month11","year1month12",
          "year2month1","year2month2","year2month3","year2month4","year2month5","year2month6"
          "year2month7","year2month8","year2month9","year2month10","year2month11","year2month12"]
```

```python
market = "BABA"
full_market = [] #Lista que almacena todos los datos de cada mercado

for sli in slices:
  #Inicia el codigo de importacion en forma de lista
  CSV_URL = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY_EXTENDED&symbol='+market+'&interval=5min&slice='+sli+'&apikey='+ API_KEY
  with requests.Session() as s:
      download = s.get(CSV_URL)
      decoded_content = download.content.decode('utf-8')
      cr = csv.reader(decoded_content.splitlines(), delimiter=',')
      my_list = list(cr)
  #Finaliza la importacion guardando los datos del mercado market, correspondiente
  #al periodo sli en forma de lista en la variable "my_list"

  full_market.extend(my_list[1:]) #El primer elemento son las etiquetas (metadata)
  
df = pd.DataFrame(full_market,columns=['time', 'open', 'high', 'low', 'close', 'volume'])
df.to_csv("Datasets/Full History/"+market+".csv",index=False)
```
